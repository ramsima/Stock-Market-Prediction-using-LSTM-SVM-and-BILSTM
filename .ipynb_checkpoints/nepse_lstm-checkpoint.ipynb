```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import matplotlib.pyplot as plt
```

```python
# Read the CSV file
data = pd.read_csv('nepse_data.csv')
```

```python
data
```

```python
# For Moving Average of 100 days
ma_100_days = data['Close'].rolling(100).mean()
```

```python
plt.figure(figsize=(8,6))
plt.plot(ma_100_days, 'r')
plt.plot(data['Close'], 'b')
plt.show()
```

```python
# For Moving Average of 200 days
ma_200_days = data['Close'].rolling(200).mean()
```

```python
plt.figure(figsize=(8,6))
plt.plot(ma_100_days, 'r')
plt.plot(data['Close'], 'b')
plt.plot(ma_200_days, 'g')
plt.show()
```

```python
### Delete all the null values in nepse data
data.dropna(inplace=True)
```

```python
# Convert the Date column to a datetime object
data['Date'] = pd.to_datetime(data['Date'])
```

```python
# Sort the dataframe by date
data = data.sort_values('Date')
```

```python
# Convert '--' to 0 in the 'Percent Change' column
data['Percent Change'] = data['Percent Change'].replace('--', 0)
```

```python
# Convert columns to float
data['Open'] = data['Open'].astype(float)
data['High'] = data['High'].astype(float)
data['Low'] = data['Low'].astype(float)
data['Close'] = data['Close'].astype(float)
data = data['Close']
```

```python
##splitting dataset into train and test split
training_size = int(len(data)*0.7)
test_size=len(data)-training_size
data_train = pd.DataFrame(data[0:training_size])
data_test = pd.DataFrame(data[training_size:len(data)])
```

```python
data_train
```

```python
data_test
```

```python
plt.plot(data)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
data_train_scale = scaler.fit_transform(data_train)
```

```python
x = []
y = []

for i in range(100, data_train_scale.shape[0]):
    x.append(data_train_scale[i-100:i])
    y.append(data_train_scale[i,0])
```

```python
x, y = np.array(x), np.array(y)
```

```python
model = Sequential()
model.add(LSTM(units = 50, activation = 'relu', return_sequences = True,
               input_shape = ((x.shape[1],1))))
model.add(Dropout(0.2))

model.add(LSTM(units = 60, activation='relu', return_sequences = True))
model.add(Dropout(0.3))

model.add(LSTM(units = 80, activation = 'relu', return_sequences = True))
model.add(Dropout(0.4))

model.add(LSTM(units = 120, activation = 'relu'))
model.add(Dropout(0.5))

model.add(Dense(units =1))
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
model.fit(x,y, epochs = 50, batch_size =32, verbose =1)
```

```python
model.summary()
```

```python
pas_100_days = data_train.tail(100)
```

```python
data_test = pd.concat([pas_100_days, data_test], ignore_index=True)
```

```python
data_test_scale  =  scaler.fit_transform(data_test)
```

```python
x = []
y = []

for i in range(100, data_test_scale.shape[0]):
    x.append(data_test_scale[i-100:i])
    y.append(data_test_scale[i,0])
x, y = np.array(x), np.array(y)
```

```python
y_predict = model.predict(x)
```

```python
scale =1/scaler.scale_
```

```python
y_predict = y_predict*scale
```

```python
y = y*scale
```

```python
plt.figure(figsize=(10,8))
plt.plot(y_predict, 'r', label = 'Predicted Price')
plt.plot(y, 'g', label = 'Original Price')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
model.save('Stock Predictions Model.keras')
```
